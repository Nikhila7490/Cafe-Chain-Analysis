```python
from apyori import apriori
```

```python
df.head()
```

```python
df1=df[['Bill_Number_','Date','Item_Desc']].copy()
```

```python
df1
```

```python
transactions = df1.groupby(['Bill_Number_','Date'])
transactions.count()
```

```python
list_transactions = [i[1]['Item_Desc'].tolist() for i in list(transactions)]
list_transactions[:20]
```

```python
rules = apriori(list_transactions, min_support=0.001, min_confidence=0.05, min_lift=1.2, min_length=2, max_length=2)
```

```python
results = list(rules)
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1]*100 for result in results]
    confidence = [result[2][0][2]*100 for result in results]
    lift = [result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lift))
final_result = pd.DataFrame(inspect(results), columns=['Product 1','Product 2','Support(%)','Confidence(%)','lift'])
final_result['Rule'] = final_result['Product 1'] + '->' + final_result['Product 2']
```

```python
final_result.head(10)
```

```python
# acc to support ascending order

final_result_1=final_result.sort_values("Support(%)")
final_result_1.head(10)
```

```python
# acc to confidence ascending order

final_result_2=final_result.sort_values("Confidence(%)")
final_result_2.head(10)
```

```python
# acc to lift ascending order

final_result_3=final_result.sort_values("lift")
final_result_3.head(10)
```

# price Analysis 

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Seperate our Pandas data frame into a dependant and independant variables
#Independant variable
X =df[['Rate']]
#Dependant variable
y = df['Quantity']
```

```python
#Perform the train and test split on our dependant and independant variables
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Instantiate our linear regression
lr = LinearRegression()
#Fit the model to our training data
lr.fit(X_train,y_train)
#Make predictions 
train_predictions = lr.predict(X_train)
test_predictions = lr.predict(X_test)
```

```python
#Import evaluation packages
from sklearn.metrics import mean_squared_error, r2_score
print(f' Train RMSE {round(mean_squared_error(y_train,train_predictions,squared=False),3)}')
print(f' Train R2 {round(r2_score(y_train,train_predictions),3)}')
print()
print(f' Test RMSE {round(mean_squared_error(y_test,test_predictions,squared=False),3)}')
print(f' Test R2 {round(r2_score(y_test,test_predictions),3)}')
print()
print(f'Price coefficient {round(lr.coef_[0],3)}')
```

```python
#Import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
#Create our scatterplot
fig, ax=plt.subplots(figsize=(10,5))
plt.scatter(y, X)
plt.plot(lr.predict(X),X,color='k')
ax.set_title("PED W/ Linear Regression")
ax.set_xlabel('Rate')
ax.set_ylabel('Quantity')
plt.show()
```
